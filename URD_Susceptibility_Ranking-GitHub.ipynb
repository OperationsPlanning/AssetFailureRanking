```python
## Import required python libraries
import lifelines
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
from lifelines import KaplanMeierFitter
from matplotlib.lines import Line2D
```

```python
## Import dataset
df = pd.read_csv('Fuse_Dataset_DF_v2.csv', index_col=[0])
```

```python
df.head(1)
```

```python
kmf_1 = KaplanMeierFitter()

T = df['Age_at_first_Fault']
C = df['Failed']

kmf_1.fit(T,C)
```

```python
%pylab inline
figsize(12,6)
kmf_2 = KaplanMeierFitter()
ax = subplot(111)

for r in df['Injected'].unique()[:]:
    loc = df['Injected'] == r
    kmf_2.fit(T.loc[loc], C.loc[loc], label = r)
    plt.title("The Kaplan-Meier Survival Estimate Injected vs Non Injected")
    ax = kmf_2.plot(ax = ax)
```

```python
## Non Injected cable survival function
kmf_noninj = KaplanMeierFitter()
for r in df['Injected'].unique()[:1]:
    loc = df['Injected'] == r
    kmf_noninj.fit(T.loc[loc], C.loc[loc], label = r)

kmf_noninj_func = kmf_2.survival_function_
```

```python
## Injected cable survival function
kmf_inj = KaplanMeierFitter()
for r in df['Injected'].unique()[1:]:
    loc = df['Injected'] == r
    kmf_inj.fit(T.loc[loc], C.loc[loc], label = r)
    
kmf_inj_func = kmf_2.survival_function_
```

```python
%pylab inline
figsize(12,6)
kmf_length = KaplanMeierFitter()
ax = subplot(111)

for r in df['1km bins '].unique()[:]:
    loc = df['1km bins '] == r
    kmf_length.fit(T.loc[loc], C.loc[loc], label = r)
    plt.title("The Kaplan-Meier Survival Estimate Length")
    ax = kmf_length.plot(ax = ax)
```
